# Scatter Plot

```python
df_total = pd.DataFrame(df[years].sum(axis=0))

df_total.index = map(int, df_total.index)

df_total.reset_index(inplace = True)
df_total.columns = ['year', 'Total']

df_total.head()
```

```python
x = df_total['year']     
y = df_total['Total']     
fit = np.polyfit(x, y, deg=1)

fit
```

```python
df_total.plot(kind='scatter', x='year', y='Total', figsize=(10, 6), color='darkblue')

plt.title('Total Immigration to US from 2011 - 2020')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')


plt.plot(x, fit[0] * x + fit[1], color='red') 
plt.annotate('y={0:.0f} x + {1:.0f}'.format(fit[0], fit[1]), xy=(2000, 150000))

plt.show()

# print out the line of best fit
'No. Immigrants = {0:.0f} * Year + {1:.0f}'.format(fit[0], fit[1]) 
```

# Bubble Plots

```python
# transposed dataframe
df_transpose = df[years].transpose()

df_transpose.index = map(int, df_transpose.index)
df_transpose.index.name = 'Year'

df_transpose.reset_index(inplace=True)
df_transpose.head()
```

```python
# normalize Eritrea data
norm_Eritrea = (df_transpose['Eritrea'] - df_transpose['Eritrea'].min()) / (df_transpose['Eritrea'].max() - df_transpose['Eritrea'].min())

# normalize Ethiopia data
norm_Ethiopia = (df_transpose['Ethiopia'] - df_transpose['Ethiopia'].min()) / (df_transpose['Ethiopia'].max() - df_transpose['Ethiopia'].min())
```

```python
# Eritrea
ax0 = df_transpose.plot(kind='scatter',
                    x='Year',
                    y='Eritrea',
                    figsize=(14, 8),
                    alpha=0.5,  # transparency
                    color='green',
                    s=norm_Eritrea * 2000 + 10, 
                    xlim=(2008, 2022)
                    )

# Ethiopia
ax1 = df_transpose.plot(kind='scatter',
                    x='Year',
                    y='Ethiopia',
                    alpha=0.5,
                    color="blue",
                    s=norm_Ethiopia * 2000 + 10,
                    ax=ax0
                    )

ax0.set_ylabel('Number of Immigrants')
ax0.set_title('Immigration from Eritrea and Ethiopia from 2011 to 2020')
ax0.legend(['Eritrea', 'Ethiopia'], loc='upper left', fontsize='x-large')
```
